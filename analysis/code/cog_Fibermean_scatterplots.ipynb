```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
#open the df you want to use
PC1_great=pd.read_csv('~/Google Drive/WIN_Matter/analysis/data/PC1_greater_sigOnly_fibers_grubbs.csv')
PC1_less= pd.read_csv('~/Google Drive/WIN_Matter/analysis/data/PC1_lesser_sigOnly_fibers_grubbs.csv')
PC2_great= pd.read_csv('~/Google Drive/WIN_Matter/analysis/data/PC2_greater_sigOnly_fibers_grubbs.csv')
PC2_less= pd.read_csv('~/Google Drive/WIN_Matter/analysis/data/PC2_lesser_sigOnly_fibers_grubbs.csv')
```

```python
def cogScatterplots(df, savename='PC'):
    IGTp=sns.jointplot(x='fiberMean', y='IGT_p_diff', data=df)
    IGTp.annotate(stats.pearsonr)
    fig = IGTp.fig 
    fig.suptitle("Mean Fiber PC2(+) By Iowa Gambling Task p-score", y=1)
    IGTp.set_axis_labels(xlabel='$\Delta Mean Fiber_{post-pre}$', ylabel='$\Delta p score_{post-pre}$')
    IGTp.savefig('/Users/aporter1350/Google Drive/WIN_Matter/analysis/output/images/'+savename+'_IGTp.png', bbox_inches='tight')
    
    IGTq=sns.jointplot(x='fiberMean', y='IGT_q_diff', data=df)
    IGTq.annotate(stats.pearsonr)
    fig = IGTq.fig 
    fig.suptitle("Mean Fiber PC2(+) By Iowa Gambling Task q-score", y=1)
    IGTq.set_axis_labels(xlabel='$\Delta Mean Fiber_{post-pre}$', ylabel='$\Delta q score_{post-pre}$')
    IGTq.savefig('/Users/aporter1350/Google Drive/WIN_Matter/analysis/output/images/'+savename+'_IGTq.png', bbox_inches='tight')
    
    TS_RT=sns.jointplot(x='fiberMean', y='TS_RT_diff', data=df)
    TS_RT.annotate(stats.pearsonr)
    fig = TS_RT.fig 
    fig.suptitle("Mean Fiber PC2(+) By Task Switch Reaction Time", y=1)
    TS_RT.set_axis_labels(xlabel='$\Delta Mean Fiber_{post-pre}$', ylabel='$\Delta RT_{post-pre}$')
    TS_RT.savefig('/Users/aporter1350/Google Drive/WIN_Matter/analysis/output/images/'+savename+'_TS_RT.png', bbox_inches='tight')
    
    TS_ACC=sns.jointplot(x='fiberMean', y='TS_ACC_diff', data=df)
    TS_ACC.annotate(stats.pearsonr)
    fig = TS_ACC.fig 
    fig.suptitle("Mean Fiber PC2(+) By Task Switch Accuracy", y=1)
    TS_ACC.set_axis_labels(xlabel='$\Delta Mean Fiber_{post-pre}$', ylabel='$\Delta ACC_{post-pre}$')
    TS_ACC.savefig('/Users/aporter1350/Google Drive/WIN_Matter/analysis/output/images/'+savename+'_TS_ACC.png', bbox_inches='tight')
    
    nback_RT=sns.jointplot(x='fiberMean', y='nback_RT_diff', data=df)
    nback_RT.annotate(stats.pearsonr)
    fig = nback_RT.fig 
    fig.suptitle("Mean Fiber PC2(+) By Nback Reaction Time", y=1)
    nback_RT.set_axis_labels(xlabel='$\Delta Mean Fiber_{post-pre}$', ylabel='$\Delta RT_{post-pre}$')
    nback_RT.savefig('/Users/aporter1350/Google Drive/WIN_Matter/analysis/output/images/'+savename+'_nback_RT.png', bbox_inches='tight')
    
    nback_ACC=sns.jointplot(x='fiberMean', y='nback_ACC_diff', data=df)
    nback_ACC.annotate(stats.pearsonr)
    fig = nback_ACC.fig 
    fig.suptitle("Mean Fiber PC2(+) By Nback Accuracy", y=1)
    nback_ACC.set_axis_labels(xlabel='$\Delta Mean Fiber_{post-pre}$', ylabel='$\Delta ACC_{post-pre}$')
    nback_ACC.savefig('/Users/aporter1350/Google Drive/WIN_Matter/analysis/output/images/'+savename+'_nback_ACC.png', bbox_inches='tight')
    
    stroop_int=sns.jointplot(x='fiberMean', y='stroop_int_diff', data=df)
    stroop_int.annotate(stats.pearsonr)
    fig = stroop_int.fig 
    fig.suptitle("Mean Fiber PC2(+) By Stroop Interference Score", y=1)
    stroop_int.set_axis_labels(xlabel='$\Delta Mean Fiber_{post-pre}$', ylabel='$\Delta Interference_{post-pre}$')
    stroop_int.savefig('/Users/aporter1350/Google Drive/WIN_Matter/analysis/output/images/'+savename+'_stroop_int.png', bbox_inches='tight')
    
    stroop_fac=sns.jointplot(x='fiberMean', y='stroop_fac_diff', data=df)
    stroop_fac.annotate(stats.pearsonr)
    fig = stroop_fac.fig 
    fig.suptitle("Mean Fiber PC2(+) By Stroop Facilitation Score", y=1)
    stroop_fac.set_axis_labels(xlabel='$\Delta Mean Fiber_{post-pre}$', ylabel='$\Delta Facilitation_{post-pre}$')
    stroop_fac.savefig('/Users/aporter1350/Google Drive/WIN_Matter/analysis/output/images/'+savename+'_stroop_fac.png', bbox_inches='tight')
```
