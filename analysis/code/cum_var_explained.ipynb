```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from jupyterthemes import jtplot

df=pd.read_csv('/data/dataDB/WIN_Matter/analysis/data/WINPCpercentvar.csv')
```

```python
#setting up colors for each health variable
#PC1: blue
#PC2: purple
#PC3: light purple
#PC4-6: green
pc=df.PC.unique().tolist()
pc_colors=["#236AB9","#5A18C9", "#66B032", "#92B6C4", "#92B6C4", "#92B6C4"]
pc_dict=dict(zip(pc, pc_colors))

sns.set_context=("talk")
plt.ylim(0, 100)
sns.set(rc={'figure.figsize':(14,10)})
ax=sns.barplot(x="PC", y="var", data=df, palette=pc_dict)
ax.set_xlabel('Principle Component Variables', fontsize=20)
ax.set_ylabel("Cumulative Variance Explained(%)",fontsize=20)
ax.tick_params(labelsize=16)
plt.yticks([0, 25, 50, 75, 100])
plt.axhline(95, linestyle ='--')
ax.get_figure().savefig("/data/dataDB/WIN_Matter/analysis/output/images/pcvar_cumplot.png", bbox_inches='tight')
```
